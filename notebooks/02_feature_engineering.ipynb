### **Feature Engineering**

```python
# total visits in previous year

def total_visits_in_previous_year(X: pd.DataFrame) -> pd.DataFrame:
    X['total_visits_in_previous_year'] = X['number_outpatient']\
                                        + X['number_emergency']\
                                        + X['number_inpatient']
    return X


train = total_visits_in_previous_year(train)
```
